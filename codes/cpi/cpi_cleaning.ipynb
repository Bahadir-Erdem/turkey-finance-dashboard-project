```python
import pandas as pd
from tabula import read_pdf
```

```python
link = 'TCMB - Tüketici Fiyatları.pdf'
pages = read_pdf(link, pages='all', pandas_options={'header': None})
```

```python
pages[1]
```

```python
data = pd.concat(pages, axis=0)
```

```python
data = data.loc[:, [0, 1]]
```

```python
data[0] = pd.to_datetime(data[0], format='%m-%Y')
```

```python
data = data.iloc[1 :, :]
```

```python
data = data.reset_index(drop=True)
```

```python
data[1] = data.loc[:, 1].astype(float)
```
