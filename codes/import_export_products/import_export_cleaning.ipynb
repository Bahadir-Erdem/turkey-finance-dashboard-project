```python
import pandas as pd
```

```python
# because of the file sizes, this repository does not contain files below
# you can download the file from this link http://www.cepii.fr/DATA_DOWNLOAD/baci/data/BACI_HS02_V202301.zip
# after the download completed you need to extract the downloaded files and make sure addresses are correct

address = './BACI_HS02_V202301/product_codes_HS02_V202301.csv'
product_codes = pd.read_csv(address)
address = './BACI_HS02_V202301/country_codes_V202301.csv'
country_codes = pd.read_csv(address)
```

```python
START_DATE = 2002
END_DATE = 2021
TURKEY_COUNTRY_CODE = 792
dataframes = []
```

```python
for year in range(START_DATE, END_DATE + 1):
    link = f'./BACI_HS02_V202301/BACI_HS02_Y{year}_V202301.csv'
    file = pd.read_csv(link)
    df = file.query('i == @TURKEY_COUNTRY_CODE or j == @TURKEY_COUNTRY_CODE')
    dataframes.append(df)
```

```python
data = pd.concat(dataframes, axis=0).reset_index(drop=True)
```

```python
data['h2_code'] =  data['k'].apply(lambda x: int(round(x / 10000)))
```

```python
data = data.merge(categories, how='inner', left_on='h2_code', right_on='code').iloc[:, [0, 1, 2, 3, 4, 5, 8]]
```

```python
turkey_import = data.query('i != @TURKEY_COUNTRY_CODE')
```

```python
turkey_export = data.query('i == @TURKEY_COUNTRY_CODE')
```

```python
turkey_import
```

```python
turkey_export
```

```python
turkey_export = turkey_export.merge(country_codes, how='inner', left_on='j', right_on='country_code').iloc[:,[0, 1, 2, 3, 4, 5, 6, 8]]
```

```python
turkey_import = turkey_import.merge(country_codes, how='inner', left_on='i', right_on='country_code').iloc[:,[0, 1, 2, 3, 4, 5, 6, 8]]
```

```python
turkey_import.loc[:, 't'] = pd.to_datetime(turkey_import['t'], format='%Y')
```

```python
turkey_export.loc[:, 't'] = pd.to_datetime(turkey_export['t'], format='%Y')
```

```python
turkey_export = turkey_export.loc[:, ~turkey_export.columns.isin(['i', 'k', 'j'])]
```

```python
turkey_import = turkey_import.loc[:, ~turkey_import.columns.isin(['k', 'j', 'i'])]
```

```python
turkey_export.columns = ['Year', 'Value of the trade flow (in thousands current USD)', 'Quantity (tons)', 'Product', 'Country']
```

```python
turkey_import.columns = ['Year', 'Value of the trade flow (in thousands current USD)', 'Quantity (tons)', 'Product', 'Country']
```
